# Changing datatypes of Height and Weight Columns 
Converting to int from str

```python
def convert_height(x):
    parts = x.split("'")
    feet = int(parts[0]) if len(parts) > 0 else 0
    inches = int(parts[1].split("\"")[0]) if len(parts) > 1 else 0
    return feet * 12 + inches

df2['Height'] = df2['Height'].astype(str).apply(convert_height)
```

```python
df2['Weight'] = df2['Weight'].astype(str).apply(lambda x: int(x.replace("lbs", "")))
```

# Seperating date to year month and Day 

```python
df2['Joined'] = pd.to_datetime(df2['Joined'])
df2['Year'] = df2['Joined'].dt.year
df2['Month'] = df2['Joined'].dt.month
df2['Day'] = df2['Joined'].dt.day
```

# Cleaning Currency column 

```python
def clean_currency(x):
    if isinstance(x, str):
        x = x.replace('€', '').replace('K', 'e3').replace('M', 'e6').replace('B', 'e9')
        return int(float(x))
    return x

df['Value'] = df['Value'].apply(clean_currency)
df['Wage'] = df['Wage'].apply(clean_currency)
df['Release Clause'] = df['Release Clause'].apply(clean_currency)

df2['Value'] = df2['Value'].apply(clean_currency)
df2['Wage'] = df2['Wage'].apply(clean_currency)
df2['Release Clause'] = df2['Release Clause'].apply(clean_currency)
```

# Removing new line character from Hits column

```python
df['Hits'] = df['Hits'].str.replace('\n', '')
df2['Hits'] = df2['Hits'].str.replace('\n', '')
```

# Seperating 'Team & Contract' column to 'Team' and 'Contract' columns

```python
df2[['Team', 'Contract']] = df2['Team & Contract'].str.extract(r'^(.*) / (.*)$')
```

```python
df2 = df2.drop('Team & Contract', axis=1)
```
